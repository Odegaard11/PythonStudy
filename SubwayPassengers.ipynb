### 1. 역별 연간 총 승객 변화율 (3년)


```python
sum1819 = subway1819[['역명', '승차총승객수']].groupby('역명').sum().sort_values(by='역명')
sum1920 = subway1920[['역명', '승차총승객수']].groupby('역명').sum().sort_values(by='역명')
sum2021 = subway2021[['역명', '승차총승객수']].groupby('역명').sum().sort_values(by='역명')
```

```python
sum2021['1819'] = sum1819['승차총승객수']
sum2021['1920'] = sum1920['승차총승객수']
sum2021['2021'] = sum2021['승차총승객수']
```

```python
total = sum2021[['1819', '1920', '2021']]
```

```python
total['diff1'] = total['1920'] / total['1819']
total['diff2'] = total['2021'] / total['1920']
```

```python
total[['1819', '1920', 'diff1']].sort_values(by='diff1', ascending=False)[6:56]
```

```python
total[['1920', '2021', 'diff2']].sort_values(by='diff2', ascending=False)[-68:-8]
```

```python
total[['1920', '2021', 'diff2']].sort_values(by='diff2', ascending=False)[:60]
```

### 2. 역별 코로나 3차 유행 영향 (3달)

```python
h10 = subway2021.loc[subway2021['year'] == '20'].loc[subway2021.loc[subway2021['year'] == '20']['month'] == '10']
h11 = subway2021.loc[subway2021['year'] == '20'].loc[subway2021.loc[subway2021['year'] == '20']['month'] == '11']
h12 = subway2021.loc[subway2021['year'] == '20'].loc[subway2021.loc[subway2021['year'] == '20']['month'] == '12']
```

```python
hollow = pd.DataFrame()
hollow['10'] = h10.groupby('역명')['승차총승객수'].sum()
hollow['11'] = h11.groupby('역명')['승차총승객수'].sum()
hollow['12'] = h12.groupby('역명')['승차총승객수'].sum()
hollow['11/10'] = hollow['11']/hollow['10']
hollow['12/11'] = hollow['12']/hollow['11']
hollow
```
